```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
#change color,size of dots in graph
colours=np.random.rand(50)
sizes=100*np.random.rand(50)
plt.scatter(x,y, c=colours,s=sizes,alpha=.7)
plt.xlabel("xaxis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
#categorical and numerical data
x=["a","b","c","d","e"]
```

```python
y=np.random.rand(5)
plt.bar(x,y)
plt.xlabel("x data")
plt.ylabel("y data")
```

```python
plt.barh(x,y)
plt.xlabel("x data")
plt.ylabel("y data")
```

```python
plt.plot(x,y,"--b")
plt.figure(figsize=(6,2))
plt.xlabel("this is my x daata")
plt.ylabel("this is my y data")
plt.title("title")
plt.show()
plt.grid()
```
